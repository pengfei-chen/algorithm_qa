####  ++i :   i+1  

####  i++ :   i + 1

```python
s = input()
m = input()


def getNextArray(ms):
    if len(ms)==1:
        return [-1]
    Next = [-1] * len(ms)
    Next[0] = -1 
    Next[1] = 0 
    pos = 2
    cn = 0 
    while pos < len(Next): # ms
        if ms[pos-1] == ms[cn]:
            cn += 1  # 
            Next[pos] = cn 
            pos += 1
            
        elif cn > 0:
            cn = Next[cn] # 0cn+1
        else:
            Next[pos] = 0 
            pos += 1
    return Next 

def getIndexOf(s,m):
    res = []
    if not s or not m or len(m)<1 or len(s) < len(m):
        return -1
    ss = list(s)
    ms = list(m)
    si,mi = 0,0  # s  m 
    Next = getNextArray(ms)
    while si < len(ss):
        mi = 0
        while (si < len(ss) and mi < len(ms)):
            if ss[si] == ms[mi]:
                si += 1
                mi += 1
            elif  Next[mi] == -1: # 
                si += 1
            else:
                mi = Next[mi]  #  
        # s s[i]
        an = (si - mi) if mi == len(ms) else -1  
        if (an != -1) or (len(res)==0 and an==-1) :
            res.append(an)
    return res

ans = getIndexOf(s,m)
print(*ans)
```

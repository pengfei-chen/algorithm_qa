###  吸取教训

####  ++i : 先 把 i+1 再进行后面的运算 

####  i++ : 先进行当前的运算，然后再 把 i + 1

```python
s = input()
m = input()


def getNextArray(ms):
    if len(ms)==1:
        return [-1]
    Next = [-1] * len(ms)
    Next[0] = -1 
    Next[1] = 0 
    pos = 2
    cn = 0 
    while pos < len(Next): # 找完ms长度
        if ms[pos-1] == ms[cn]:
            cn += 1  # 万万没想到，是这里让我一直卡着。
            Next[pos] = cn 
            pos += 1
            
        elif cn > 0:
            cn = Next[cn] # 索引从0开始，此处就等于长度为cn+1的索引
        else:
            Next[pos] = 0 
            pos += 1
    return Next 

def getIndexOf(s,m):
    res = []
    if not s or not m or len(m)<1 or len(s) < len(m):
        return -1
    ss = list(s)
    ms = list(m)
    si,mi = 0,0  # s 和 m 的匹配位置
    Next = getNextArray(ms)
    while si < len(ss):
        mi = 0
        while (si < len(ss) and mi < len(ms)):
            if ss[si] == ms[mi]:
                si += 1
                mi += 1
            elif  Next[mi] == -1: # 这里待会看一次
                si += 1
            else:
                mi = Next[mi]  # 加速主要看这里 
        # s走到 s[i]这里，就已经匹配完了
        an = (si - mi) if mi == len(ms) else -1  
        if (an != -1) or (len(res)==0 and an==-1) :
            res.append(an)
    return res

ans = getIndexOf(s,m)
print(*ans)
```

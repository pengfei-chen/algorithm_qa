###  写法学习

```python
str1 = input()
str2 = input()

def minLength(str1, str2):
    if not str1 or not str2 or len(str1)==0 or len(str2)==0:
        return 0
    chas1 = list(str1)
    chas2 = list(str2)
    Map = dict()
    for i in range(len(chas2)): #str2里面的字符串个数统计
        Map[chas2[i]] = Map.get(chas2[i], 0)  + 1 
    left,right = 0,0 
    match = len(chas2)
    minLen = 10**10 
    while right != len(chas1): # right还没到终点
        Map[chas1[right]] = Map.get(chas1[right], 0) - 1 # str1中的字符减1
        if Map[chas1[right]] >= 0: #说明之前是>=1的，现在 还回 一个 
            match -= 1
        if match == 0:
            while Map[chas1[left]] < 0: #left从0开始往右走
                Map[chas1[left]] = Map.get(chas1[left],0) + 1 
                left += 1
            minLen = min(minLen, right-left+1) # 更新一次minLen
            match += 1 # left还要往右走，
            Map[chas1[left]] = Map.get(chas1[left],0) + 1 
            left += 1
        right += 1
    return 0 if minLen == 10**10 else minLen 

print(minLength(str1, str2))
```

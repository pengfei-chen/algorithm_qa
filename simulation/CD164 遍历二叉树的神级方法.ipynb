```python
class Node:
    def __init__(self,x):
        self.v = x 
        self.left = None 
        self.right = None 
        
# 根据输入构造二叉树 
n,r = input().split(' ')
root = Node(r)
node_dict = dict()
node_dict[r] = root 
for i in range(int(n)):
    fa,lch,rch = input().split(' ')
    if lch != '0':
        node_dict[lch] = Node(lch)
        node_dict[fa].left = node_dict[lch]
    if rch != '0':
        node_dict[rch] = Node(rch)
        node_dict[fa].right = node_dict[rch]
        
# 懒了一点。。
def morrisPre(root):
    ans = []
    node = root
    while node:
        if node.left:
            mostR = node.left
            while mostR.right and mostR.right != node:
                mostR = mostR.right
            if not mostR.right:
                mostR.right = node
                ans.append(node.v)
                node = node.left
            else:
                mostR.right = None
                node = node.right
        else:
            ans.append(node.v)
            node = node.right
    return ans
 
def morrisIn(root):
    ans = []
    node = root
    while node:
        if node.left:
            mostR = node.left
            while mostR.right and mostR.right != node:
                mostR = mostR.right
            if not mostR.right:
                mostR.right = node
                node = node.left
            else:
                mostR.right = None
                ans.append(node.v)
                node = node.right
        else:
            ans.append(node.v)
            node = node.right
    return ans
 
def reverse(head):
    pre = None
    while head:
        next = head.right
        head.right = pre
        pre = head
        head = next
    return pre
 
def morrisPos(root):
    def getEdge(head):
        tail = reverse(head)
        node = tail
        while node:
            ans.append(node.v)
            node = node.right
        reverse(tail)
    ans = []
    node = root
    while node:
        if node.left:
            mostR = node.left
            while mostR.right and mostR.right != node:
                mostR = mostR.right
            if not mostR.right:
                mostR.right = node
                node = node.left
            else:
                mostR.right = None
                getEdge(node.left)
                node = node.right
        else:
            node = node.right
    getEdge(root)
    return ans

pre_order = morrisPre(root)
print(' '.join(pre_order))
in_order = morrisIn(root)
print(' '.join(in_order))
pos_order = morrisPos(root)
print(' '.join(pos_order))
```

```python
n = int(input())
arr = []
for i in range(n):
    lst = list(map(int,input().split()))
    arr.append(lst)

def func(arr):
    n = len(arr)
    if not n:
        return 0
    arr.sort(key = lambda x : (x[0], -x[1]))
    top = [0] * n
    piles = 0
    for env in arr:
#         w, h = env[0], env[1]
        h = env[1]
        left , right = 0, piles
        while left != right:
            mid = (left + right)//2
            if top[mid] < h:
                left = mid+1
            else:
                right = mid
        if left == piles:
            piles += 1
        top[left] = h
    return piles
print(func(arr))
```

```python
n = int(input())
arr = []
for i in range(n):
    lst = list(map(int,input().split()))
    arr.append(lst)

def func(arr):
    n = len(arr)
    if not n:
        return 0
    arr.sort(key = lambda x : (x[0], x[1]))
    Max = 0
    for i in range(n):
        l = i
        cnt = 1
        while i < n-1:
            while not (arr[l][0] < arr[i][0]):
                i += 1
            if arr[l][1] < arr[i][1]:
                cnt += 1
                l = i
            i += 1
        if cnt > Max:
            Max = cnt
    return Max
print(func(arr))
```

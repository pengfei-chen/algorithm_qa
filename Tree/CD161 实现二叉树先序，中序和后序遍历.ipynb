### 使用非递归的方式遍历二叉树！

```python
# 定义树的类
class Node:
    def __init__(self, x=None):
        self.v = x 
        self.left = None 
        self.right = None 
    
n,root = list(map(int,input().split()))
nodes = [Node(i) for i in range(n+1)]
nodes[0] = None 

for i in range(n):
    fa,lch,rch = list(map(int,input().split())) 
    nodes[fa].left = nodes[lch] 
    nodes[fa].right = nodes[rch] 
    
head = nodes[root]
# nodes[root]   #根节点 nodes[root]
# 递归方式
class solution:
    def __init__(self):
        self.ans = []

    # 递归方式先序遍历
    def preOrderRecur(self,node):
        if node is None:
            return 
        self.ans.append(node.v)
        self.preOrderRecur(node.left)
        self.preOrderRecur(node.right)   
    # 非递归方式先序遍历
    def preOrderUnRecur(self,node):
        if node is not None:
            stack = []
            stack.append(node)
            while len(stack) != 0:
                head = stack.pop()
                self.ans.append(head.v)
                # 先入右子节点，先入后出
                if head.right is not None:
                    stack.append(head.right)
                if head.left is not None:
                    stack.append(head.left) 
                    
    # 递归方式中序遍历
    def midOrderRecur(self, node):
        if node is None:
            return 
        self.midOrderRecur(node.left)
        self.ans.append(node.v)
        self.midOrderRecur(node.right)
        
    # 非递归方式中序遍历
    def midOrderUnRecur(self, node):
        if node is not None:
            stack = []
            while (len(stack) != 0) or (node is not None):
                if node is not None:
                    stack.append(node)
                    node = node.left # 一直往左边走
                else:
                    node = stack.pop() # 弹出一个节点 
                    self.ans.append(node.v)
                    node = node.right 
                    
    # 递归方式后序遍历
    def posOrderRecur(self, node):
        if node is None:
            return 
        self.posOrderRecur(node.left)
        self.posOrderRecur(node.right)
        self.ans.append(node.v)
    # 非递归方式后序遍历
    def posOrderUnRecur(self, node):
        if node is not None:
            s1 = []
            s2 = []
            s1.append(node)
            while len(s1) != 0:
                node = s1.pop()
                s2.append(node.v) #根节点最先一个进入，最后一个出来
                # 用了两个栈，先进入的就是先出来的 
                if node.left is not None:
                    s1.append(node.left)
                if node.right is not None:
                    s1.append(node.right) 
            # python里面栈结构再导出，直接逆序就可以了
            s2 = s2[::-1]
            self.ans = s2
        
    def getdata(self,head):
        self.res = []
#         self.preOrderRecur(head)
        self.preOrderUnRecur(head)
        a ,self.ans = self.ans, []
        self.res.append(a)
#         self.midOrderRecur(head)
        self.midOrderUnRecur(head)
        b ,self.ans = self.ans, []
        self.res.append(b)
#         self.posOrderRecur(head)
        self.posOrderUnRecur(head)
        c ,self.ans = self.ans, []
        self.res.append(c)
        return self.res

method = solution()
ans = method.getdata(head)
for x in ans:
    print(' '.join(str(i) for i in x))
```

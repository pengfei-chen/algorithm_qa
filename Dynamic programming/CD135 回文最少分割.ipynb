```python
str1 = input()

def minCut(str1):
    if str1 is None or len(str1)==0 or len(set(str1))==1:
        return 0 
    # 如果只有一个字母，return 0 
    Len = len(str1)
    dp = [0] * (Len+1)
    dp[Len] = -1
    p = []
    for i in range(Len):
        a = [False] * Len 
        p.append(a) 
    for i in range(Len-1, -1, -1):
        dp[i] = 10**10 
        for j in range(i,Len):
            # 只有一个字母，或者str1去掉首尾i和j以后仍旧是回文串
            if (str1[i] == str1[j]) and (j-i <2 or p[i+1][j-1]):
                p[i][j] = True  # i是逐步减少的，算i时，i+1已经算好了
                dp[i] = min(dp[i], dp[j+1]+1) 
    return dp[0] 

print(minCut(str1))
```

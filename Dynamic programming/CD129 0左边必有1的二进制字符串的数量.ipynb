###  暴力方法

```python
n = int(input())

def process(i,n):
    if i == n-1:
        return 2
    if i == n:
        return 1
    return process(i+1, n) + process(i+2, n)

def getNum1(n):
    if n<1:
        return 0
    return process(1,n)

print(getNum1(n))
```

### 普通斐波那契方法

```python
n = int(input())

def getNum2(n):
    if n < 1:
        return 0 
    if n == 1:
        return 1
    pre, cur = 1,1 
    for i in range(2,n+1):
        pre,cur = cur,cur+pre
    return cur

print(getNum2(n))
```

### log2N 方法

```python
n = int(input())
MOD = pow(2, 29)

def matrix_mul(a, b):
    c0 = (a[0][0] * b[0][0] + a[1][0] * b[0][1]) % MOD
    c1 = (a[0][0] * b[1][0] + a[1][0] * b[1][1]) % MOD
    c2 = (a[0][1] * b[0][0] + a[1][1] * b[0][1]) % MOD
    c3 = (a[0][1] * b[1][0] + a[1][1] * b[1][1]) % MOD
    return [[c0, c1], [c2, c3]]

def matrix_pow(m,p):
    l = len(m)
    res = [[0] * l for _ in range(l)]
    for i in range(l):
        res[i][i] = 1 
    tmp = m 
    while p != 0:
        if p & 1 != 0:
            res = matrix_mul(res, tmp)
        tmp = matrix_mul(tmp, tmp)
        p >>= 1 
    return res 

def getNum3(n):
    if n<1:
        return 0
    if n==1 or n==1:
        return n 
    base = [[1,1],[1,0]]
    res = matrix_pow(base, n-2)
    ans = (2*res[0][0] + res[1][0]) % MOD
    return ans

print(getNum3(n))
```
